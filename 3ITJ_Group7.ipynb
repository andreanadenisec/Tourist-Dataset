```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dfPlaces = pd.read_csv('destinations.csv')
dfPlaces
```

```python
activities = dfPlaces.groupby(['Activities']).size()
activitiesDict = activities.to_dict()
activitiesDict = {key: value for key, value in sorted(activitiesDict.items(), key=lambda item: item[1], reverse=True,)}

activitiesName = list(activitiesDict.keys())
activitiesNum = list(activitiesDict.values())
colors = ['hotpink', 'blueviolet', 'dodgerblue','olivedrab', 'gold', 'darkorange', 'chocolate', 'maroon', 'rosybrown', 'lightgray']
```

```python
plt.pie(activitiesNum, autopct='%.1f%%', pctdistance=1.14, radius=1.15, colors=colors)
plt.title('Popular Types Of Tourist Destinations Around The World',fontsize=18, loc='center', y=1.08)
plt.legend(labels=activitiesName, loc='upper center', bbox_to_anchor=(0.5, -0.04), ncol=4)
plt.show()
```

```python
dfAF = dfPlaces[dfPlaces['Continent']=='Africa']
dfAS = dfPlaces[dfPlaces['Continent']=='Asia']
dfEU = dfPlaces[dfPlaces['Continent']=='Europe']
dfNA = dfPlaces[dfPlaces['Continent']=='North America']
dfOC = dfPlaces[dfPlaces['Continent']=='Oceania']
dfSA = dfPlaces[dfPlaces['Continent']=='South America']

activitiesAF = dfAF.groupby(['Activities']).size()
activitiesAS = dfAS.groupby(['Activities']).size()
activitiesEU = dfEU.groupby(['Activities']).size()
activitiesNA = dfNA.groupby(['Activities']).size()
activitiesOC = dfOC.groupby(['Activities']).size()
activitiesSA = dfSA.groupby(['Activities']).size()

activitiesDictAF = activitiesAF.to_dict()
activitiesDictAS = activitiesAS.to_dict()
activitiesDictEU = activitiesEU.to_dict()
activitiesDictNA = activitiesNA.to_dict()
activitiesDictOC = activitiesOC.to_dict()
activitiesDictSA = activitiesSA.to_dict()

activitiesDictAF = {key: value for key, value in sorted(activitiesDictAF.items(), key=lambda item: item[1], reverse=True,)}
activitiesDictAS = {key: value for key, value in sorted(activitiesDictAS.items(), key=lambda item: item[1], reverse=True,)}
activitiesDictEU = {key: value for key, value in sorted(activitiesDictEU.items(), key=lambda item: item[1], reverse=True,)}
activitiesDictNA = {key: value for key, value in sorted(activitiesDictNA.items(), key=lambda item: item[1], reverse=True,)}
activitiesDictOC = {key: value for key, value in sorted(activitiesDictOC.items(), key=lambda item: item[1], reverse=True,)}
activitiesDictSA = {key: value for key, value in sorted(activitiesDictSA.items(), key=lambda item: item[1], reverse=True,)}

activitiesNameAF = list(activitiesDictAF.keys())
activitiesNameAS = list(activitiesDictAS.keys())
activitiesNameEU = list(activitiesDictEU.keys())
activitiesNameNA = list(activitiesDictNA.keys())
activitiesNameOC = list(activitiesDictOC.keys())
activitiesNameSA = list(activitiesDictSA.keys())

activitiesNumAF = list(activitiesDictAF.values())
activitiesNumAS = list(activitiesDictAS.values())
activitiesNumEU = list(activitiesDictEU.values())
activitiesNumNA = list(activitiesDictNA.values())
activitiesNumOC = list(activitiesDictOC.values())
activitiesNumSA = list(activitiesDictSA.values())
```

```python
fig, ax = plt.subplots(2, 3)

ax[0, 1].pie(activitiesNumAF, autopct='%.1f%%', pctdistance=1.25, textprops={'fontsize': 8}, colors=colors)
ax[0, 0].pie(activitiesNumAS, autopct='%.1f%%', pctdistance=1.25, textprops={'fontsize': 8}, colors=colors)
ax[0, 2].pie(activitiesNumEU, autopct='%.1f%%', pctdistance=1.25, textprops={'fontsize': 8}, colors=colors)
ax[1, 0].pie(activitiesNumNA, autopct='%.1f%%', pctdistance=1.25, textprops={'fontsize': 8}, colors=colors)
ax[1, 1].pie(activitiesNumOC, autopct='%.1f%%', pctdistance=1.25, textprops={'fontsize': 8}, colors=colors)
ax[1, 2].pie(activitiesNumSA, autopct='%.1f%%', pctdistance=1.25, textprops={'fontsize': 8}, colors=colors)

ax[0, 1].set_title('Africa', y=1.08)
ax[0, 0].set_title('Asia', y=1.08)
ax[0, 2].set_title('Europe', y=1.08)
ax[1, 0].set_title('North America', y=1.08)
ax[1, 1].set_title('Oceania', y=1.08)
ax[1, 2].set_title('South America', y=1.08)

fig.tight_layout()
fig.suptitle('Popular Types Of Tourist Destinations Per Continent', fontsize=18, y=1.08)
fig.legend(labels=activitiesName, loc='upper center', bbox_to_anchor=(0.5, -0.04), ncol=4)

plt.show()
```
